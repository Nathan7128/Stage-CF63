```python
competitions = []

next_page = True 
page = 0
display_team_ids = 'false'
gender = 'male'

params = {
    'teamIds': display_team_ids,
    'gender': gender,
    'page': page
    }

while next_page == True:  
    response_competitions = requests.get('https://cf63.scoutastic.com/api/v1/competitions', params=params, headers=headers)
    competitions_result = response_competitions.json()
    competitions.append(competitions_result['docs'])
    next_page = competitions_result['hasNextPage']
    params['page'] += 1

competitions = list(itertools.chain(*competitions))
```
