```python
from skillcorner.client import SkillcornerClient

import pandas as pd

import numpy as np

import os

secret_password = os.getenv("mdp_skillcorner")
client = SkillcornerClient(username = "Nathan.talbot@etu.uca.fr", password = secret_password)
```

```python
dico_saison = {
    "2023_2024" : 549,
    "2022_2023" : 393,
    "2021_2022" : 243
}
```

```python
for saison in dico_saison.keys() :
    df = pd.DataFrame()
    matches_import = client.get_matches(params = {"competiton_id" : dico_saison[saison]})
    for matches in matches_import :
        match = client.get_match(matches["id"])
        if match["away_team_score"] > match["home_team_score"] :
            df.loc[matches["id"], "Résultat"] = matches["away_team"]["short_name"]
        elif match["away_team_score"] < match["home_team_score"] :
            df.loc[matches["id"], "Résultat"] = matches["home_team"]["short_name"]
    df.to_excel(f"Skill Corner/{saison}.xlsx")
```
