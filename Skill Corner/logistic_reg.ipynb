```python
# Création du Data Set
X = pd.DataFrame()


for i in range (0, len(liste_match), nb_journee*10) :
    physical_data_json = client.get_physical(params = {"data_version" : 3, "match" : liste_match[i : i + nb_journee*10], 'competition_edition': 549, "group_by" : "team",
                                                                     "possession" : ["tip", "otip"]})
    physical_data = pd.DataFrame(physical_data_json).set_index("team_name")
    physical_data.drop(["team_id", "count_match", "count_match_failed"], inplace = True, axis = 1)
    physical_data = physical_data.reindex(ranking)
    X = pd.concat([X, physical_data])
```
